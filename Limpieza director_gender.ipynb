```python
pelis["ISO_codes"] = pelis["country"].map(ISO_codes)
pelis
```

```python
#Defino una funciÃ³n para obtener el gÃ©nero de los nombres que me determina como unknown gracias a la API Genderize, utilizando
#los cÃ³digos ISO:

def get_gender(name, country_code):
    url = f"https://api.genderize.io/?name={name}&country_id={country_code}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if 'gender' in data:
            return data['gender']
    return None

# Luego recorro cada fila del Dataframe con iterrows. Si el gÃ©nero es unknown, que me aplique la funciÃ³ get_gender. 
#Si no, que lo pase a male, ya que la gran mayorÃ­a son male.

for i, row in pelis.iterrows():
    if row['gender'] == 'unknown':
        # llamar a la API Genderize.io para obtener el gÃ©nero del director
        gender = get_gender(row['director'], row['ISO_codes'])
        if gender:
            pelis.at[i, 'gender'] = gender
        else:
            pelis.at[i, 'gender'] = 'male'
```
